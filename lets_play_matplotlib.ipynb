```python
"""
VG Sales Demo: Custom Styles + Annotations + Export Quality
Downloads dataset from the provided URL, builds a styled chart, annotates a key insight,
and exports high-quality assets (PNG 150/300/600 DPI, SVG, PDF).

Requirements:
pip install pandas matplotlib requests
"""

import os
from pathlib import Path
from datetime import datetime
import requests
import io

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 0. Configuration
# -----------------------------
CSV_URL = (
    "https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/021/299/original/"
    "final_vg1_-_final_vg_%281%29.csv?1670840166"
)
LOCAL_CSV = Path("vgsales.csv")
EXPORT_DIR = Path("exports")
EXPORT_DIR.mkdir(parents=True, exist_ok=True)

# -----------------------------
# 1. Download dataset (if not present)
# -----------------------------
def download_csv(url: str, out_path: Path):
    if out_path.exists():
        print(f"File already exists locally: {out_path}")
        return out_path
    print("Downloading dataset...")
    r = requests.get(url, timeout=30)
    r.raise_for_status()
    # Write to disk
    out_path.write_bytes(r.content)
    print(f"Saved dataset to {out_path}")
    return out_path

download_csv(CSV_URL, LOCAL_CSV)

# -----------------------------
# 2. Load & clean dataset robustly
# -----------------------------
def load_vgsales(path: Path) -> pd.DataFrame:
    # The file may have inconsistent quoting/encoding. Use python engine and try multiple attempts.
    read_opts = dict(engine="python", skipinitialspace=True)
    df = pd.read_csv(path, **read_opts)
    # Standardize column names
    df.columns = [c.strip() for c in df.columns]
    # Convert numeric sales columns to floats (coerce errors)
    sales_cols = [c for c in df.columns if "Sales" in c or "Sales" in c.title()]
    for c in ["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    # Year: try to coerce to numeric (some rows may contain NaN)
    if "Year" in df.columns:
        df["Year"] = pd.to_numeric(df["Year"], errors="coerce").astype("Int64")
    # Basic cleanup: drop rows without Global_Sales
    if "Global_Sales" in df.columns:
        df = df.dropna(subset=["Global_Sales"])
    return df

df = load_vgsales(LOCAL_CSV)
print(f"Loaded {len(df)} rows, columns: {list(df.columns)}")

# Save a cleaned copy for reproducibility
clean_csv_path = EXPORT_DIR / "vgsales_clean.csv"
df.to_csv(clean_csv_path, index=False)
print("Saved cleaned dataset to:", clean_csv_path)

# -----------------------------
# 3. Aggregate/prepare data for visualization
# -----------------------------
# Example analysis: Top 10 Games by Global Sales and aggregate by Genre for trend
top_games = df.sort_values("Global_Sales", ascending=False).head(10).copy()
genre_summary = df.groupby("Genre", dropna=False)["Global_Sales"].sum().sort_values(ascending=False)
platform_summary = df.groupby("Platform", dropna=False)["Global_Sales"].sum().sort_values(ascending=False)

# Build a time series: total global sales per year (for available years)
if "Year" in df.columns:
    yearly = (
        df.dropna(subset=["Year"])
        .groupby("Year", as_index=True)["Global_Sales"]
        .sum()
        .sort_index()
        .rename("global_sales")
        .reset_index()
    )
else:
    yearly = pd.DataFrame({"Year": [], "global_sales": []})

# -----------------------------
# 4. Custom style function (reusable)
# -----------------------------
def apply_custom_style(ax):
    ax.set_facecolor("#F5F7FA")
    ax.figure.set_facecolor("white")
    # Spines
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_color("#424242")
    ax.spines["bottom"].set_color("#424242")
    # Grid
    ax.grid(which="major", linestyle="--", linewidth=0.6, color="#C5D1E0", alpha=0.9)
    ax.tick_params(axis="both", labelsize=10, colors="#333333")
    ax.title.set_fontsize(14)
    ax.title.set_weight("bold")
    ax.title.set_color("#1A237E")
    ax.xaxis.label.set_fontsize(11)
    ax.yaxis.label.set_fontsize(11)

# -----------------------------
# 5. Export function (multi-format, multi-dpi)
# -----------------------------
def export_chart(fig, filename_base, export_dir=EXPORT_DIR):
    export_dir = Path(export_dir)
    export_dir.mkdir(parents=True, exist_ok=True)
    png_150 = export_dir / f"{filename_base}_150dpi.png"
    png_300 = export_dir / f"{filename_base}_300dpi.png"
    png_600 = export_dir / f"{filename_base}_600dpi.png"
    svg_file = export_dir / f"{filename_base}.svg"
    pdf_file = export_dir / f"{filename_base}.pdf"

    fig.savefig(png_150, dpi=150, bbox_inches="tight")
    fig.savefig(png_300, dpi=300, bbox_inches="tight")
    fig.savefig(png_600, dpi=600, bbox_inches="tight")
    fig.savefig(svg_file, bbox_inches="tight")
    fig.savefig(pdf_file, bbox_inches="tight")

    print("Exported files:")
    for p in [png_150, png_300, png_600, svg_file, pdf_file]:
        print("  ", p)

# -----------------------------
# 6. Build the chart (time-series + annotation + side bar: top genres)
# -----------------------------
fig = plt.figure(constrained_layout=False, figsize=(13, 6), dpi=120)
# Use gridspec to create main (time series) and small bar chart (genres)
gs = fig.add_gridspec(1, 3, width_ratios=[2, 1, 0.8], wspace=0.22)

ax_main = fig.add_subplot(gs[0, 0:2])  # main time series
ax_side = fig.add_subplot(gs[0, 2])    # small bar: top genres

# Main: yearly global sales
apply_custom_style(ax_main)
if not yearly.empty:
    ax_main.plot(yearly["Year"], yearly["global_sales"], label="Global Sales (sum by year)",
                 linewidth=2.6, marker="o", markersize=6, color="#1976D2")
    ax_main.set_xlabel("Year")
    ax_main.set_ylabel("Global Sales (millions)")
    ax_main.set_title("Global Video Game Sales by Year (aggregated)")
    # annotate the maximum year and value
    max_row = yearly.loc[yearly["global_sales"].idxmax()]
    max_year, max_sales = int(max_row["Year"]), float(max_row["global_sales"])
    ax_main.scatter([max_year], [max_sales], s=140, facecolors='none', edgecolors="#D32F2F", linewidths=1.6, zorder=6)
    ax_main.annotate(
        f"Peak {max_year}: {max_sales:.2f}M",
        xy=(max_year, max_sales),
        xytext=(max_year - 3 if (yearly["Year"].min() < max_year - 3) else max_year + 1, max_sales * 1.12),
        arrowprops=dict(arrowstyle="->", color="#D32F2F"),
        bbox=dict(boxstyle="round,pad=0.3", fc="#FFF2F2", ec="#D32F2F"),
        fontsize=10
    )
else:
    ax_main.text(0.5, 0.5, "No Year data available", ha="center", va="center", fontsize=12)
    ax_main.set_title("Yearly Sales: No data")

# Side: top 8 genres by global sales (horizontal bar)
apply_custom_style(ax_side)
top_genres = genre_summary.head(8).sort_values(ascending=True)  # small -> large for horizontal bar
ax_side.barh(top_genres.index.astype(str), top_genres.values, height=0.65)
ax_side.set_title("Top Genres (Global Sales)")
ax_side.set_xlabel("Total Global Sales (millions)")

# Tweak layout and add watermark
timestamp = datetime.utcnow().strftime("%Y-%m-%d %H:%M UTC")
fig.text(0.99, 0.01, f"Source: VG Sales dataset (downloaded) — Generated {timestamp}",
         ha="right", fontsize=8, color="#555555", alpha=0.7)

plt.tight_layout(rect=[0, 0.02, 1, 0.98])

# -----------------------------
# 7. Export and show
# -----------------------------
export_chart(fig, "vg_sales_trend_and_genres")
plt.show()

# -----------------------------
# 8. Also produce a focused chart: Top 10 Games (annotated)
# -----------------------------
fig2, ax2 = plt.subplots(figsize=(11, 6), dpi=120)
apply_custom_style(ax2)
ax2.barh(top_games["Name"].iloc[::-1], top_games["Global_Sales"].iloc[::-1], height=0.7, color="#1976D2")
ax2.set_xlabel("Global Sales (millions)")
ax2.set_title("Top 10 Games by Global Sales (from dataset)")
# annotate the top game
top_name = top_games.iloc[0]["Name"]
top_val = top_games.iloc[0]["Global_Sales"]
ax2.annotate(
    f"Top: {top_name} — {top_val:.2f}M",
    xy=(top_val, top_games.iloc[0]["Name"]),
    xytext=(top_val + top_val * 0.08, 0),
    va="center",
    fontsize=10,
    arrowprops=dict(arrowstyle="->", color="#333333"),
    bbox=dict(boxstyle="round,pad=0.3", fc="#FFFDE7", ec="#FBC02D")
)

fig2.text(0.99, 0.01, f"Dataset cleaned: {clean_csv_path.name} — {timestamp}", ha="right", fontsize=8, alpha=0.7)
export_chart(fig2, "vg_top10_games")
plt.show()

print("\nDone. All exports are in the 'exports' folder.")
```
